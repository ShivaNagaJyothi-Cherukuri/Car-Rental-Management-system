```python
#most popular insurance plan
df_pop_ins = pd.read_sql_query("select ci.Ins_ID, ci.InsuranceName, count(*) as 'Total_Insurances_Taken' from carinsurance as ci, carbooking as cb where ci.Ins_ID = cb.Ins_ID group by ci.Ins_ID order by Total_Insurances_Taken desc limit 5", con)
df_pop_ins
```

```python
#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:10]

#create pie chart
plt.pie(df_pop_ins['Total_Insurances_Taken'], labels = df_pop_ins['InsuranceName'], colors = colors, autopct='%.0f%%')
plt.title('Most popular Insurance plan purchased by customers')
plt.show()
```

```python
#Cars available in 'Boston'
df_cars_bos = pd.read_sql_query("Select c.CarName, c.CarRegNumber from cars c  inner join location l on c.Current_LocID = l.Loc_ID  where c.Availabliity = 'Y'  and l.City = 'Boston'",con)
df_cars_bos
```

```python
#Popular car type among customers
df_carstype_pop = pd.read_sql_query("select c.CarType, count(*) as 'Total_Bookings' from carbooking as cb inner join cars as c on cb.CarRegNumber = c.CarRegNumber group by c.CarType Order by Total_Bookings desc",con)
df_carstype_pop
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=df_carstype_pop['CarType'], y=df_carstype_pop['Total_Bookings'] )
plt.title('Most Popular car types among customers')
plt.xlabel('CarTypes')
plt.ylabel('Total_Bookings')
plt.show()
```

```python
#Age of customer vs car type
df_carstype_age = pd.read_sql_query("select c.CarType, AVG(cust.Age) as 'Average Age' from carbooking as cb, cars as c, customer as cust Where cb.CarRegNumber = c.CarRegNumber and cb.CustID = cust.Cust_ID group by c.CarType",con)
df_carstype_age
```

```python
plt.plot(df_carstype_age['CarType'],df_carstype_age['Average Age'])
plt.title('Average Age of customer using different cartypes')
plt.xlabel('CarType')
plt.xticks(rotation = 45)
plt.ylabel('Average Age Group')
plt.show()
```

```python
df_noofbookings = pd.read_sql_query("SELECT MONTH(FromDate) as Month, COUNT(*) as 'Number of Bookings' FROM carbooking GROUP BY MONTH(FromDate) ORDER BY Month",con)
df_noofbookings
```

```python
# Create a line chart
plt.figure(figsize=(8,6))
sns.lineplot(x=df_noofbookings['Month'], y=df_noofbookings['Number of Bookings'])
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.title('Number of Car Bookings every month')
plt.show()
```

```python
#Histogram for Distribution of Customer Age
df_age = pd.read_sql_query("SELECT age FROM customer",con)
plt.figure(figsize=(8,6))
sns.histplot(data = df_age, x='age')
plt.xlabel('Age bins')
plt.ylabel('Number of Customers in each bin')
plt.title('Histogram for Distribution of Customer Age')
plt.show()
```

```python
df_year_price = pd.read_sql_query("SELECT cp.LateFee, cp.CostPerHour, c.carType FROM cars as c, carpricing as cp where c.CarRegNumber = cp.CarRegNumber",con)
plt.figure(figsize=(7,5))
sns.scatterplot(data = df_year_price, x='LateFee', y = 'CostPerHour', hue='carType')
plt.xlabel('LateFee')
plt.ylabel('CostPerHour')
plt.title('Scatter Plot of LateFee v/s Cost Per Hour')
plt.show()
```

```python
df_year_price = pd.read_sql_query("SELECT m.Membership_Discount, p.PlanPrice FROM membershipplans as p, membership as m where m.PlanID = p.PlanID",con)
plt.figure(figsize=(7,5))
sns.scatterplot(data = df_year_price, x='Membership_Discount', y = 'PlanPrice')
plt.xlabel('Membership_Discount')
plt.ylabel('PlanPrice')
plt.title('Scatter Plot of Membership_Discount v/s PlanPrice')
plt.show()
```

```python
df_total_cost = pd.read_sql_query("SELECT TotalCost FROM billing WHERE BillStatus = 'PAID'",con)
plt.figure(figsize=(4,6))
sns.boxplot(data = df_total_cost, y='TotalCost')
plt.xlabel('TotalCost')
plt.title('Box Plot for distribution of Totalcost paid by customers')
plt.show()
```
